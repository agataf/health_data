```python
## LDA
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
X = df.values
```

```python
X_no_names = X[:,2:]
```

```python
X_no_names[np.isnan(X_no_names.astype(np.float32))] = 0
```

```python
lda = LatentDirichletAllocation()
```

```python
Z = lda.fit_transform(X_no_names)
```

```python
plt.plot(Z[:,7],Z[:,8],'o',markersize=2)
plt.show()
```

```python
for i in range(0,10):
    plt.bar(range(0,X_no_names.shape[1]),lda.components_[i])
    plt.show()
```

```python
for i in range(0,10):
    plt.bar(range(21,X_no_names.shape[1]),lda.components_[i,21:])
plt.show()
```

```python
plt.hist(np.max(Z,axis=1))
plt.show()
```

```python
bar_width = 0.5
for i in range(0,10):
    plt.bar(10*bar_width*np.ones(642-613)*range(0,642-613) + i*bar_width*np.ones(642-613),
            lda.components_[i,613:642],
            bar_width, 
            label = str(i))
plt.legend()
plt.show()
```
